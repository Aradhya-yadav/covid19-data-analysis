```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df = covid_df.iloc[:-28]
# remove because these are NaN value
```

```python
covid_df = covid_df.dropna()
covid_df['Active_Cases'] = covid_df['Confirmed'] - covid_df['Cured'] - covid_df['Deaths']
covid_df.tail()
```

```python
statewise = pd.pivot_table(
    covid_df,
    values=["Confirmed", "Deaths", "Cured"],
    index="State/UnionTerritory",
    aggfunc="max"
)

```

```python
statewise["Recovery Rate"]=statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by="Confirmed", ascending=False)
statewise.head(10)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
top_10_active_case = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']] \
.sort_values(by=['Active_Cases'], ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title(" Top 10 States with most active cases in india",  size=25)
```

```python
top_10_active_case = covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']] \
.sort_values(by=['Active_Cases'], ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title(" Top 10 States with most active cases in india",  size=25)
ax = sns.barplot(
    data=top_10_active_case.iloc[:10],
    y="Active_Cases",
    x="State/UnionTerritory",
    linewidth=2,
    edgecolor='blue'
)
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top states with highest deaths
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']] \
    .sort_values(by='Deaths', ascending=False) \
    .reset_index()
plt.figure(figsize=(18,5))

plt.title("Top 10 States with Most Deaths", fontsize=25)
ax = sns.barplot(data=top_10_deaths.iloc[:12],
    y="Deaths",
    x="State/UnionTerritory",

    linewidth=2,
    edgecolor='black'
)
plt.xlabel("State")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
# ðŸ“ˆ Growth Trend of Top 5 Affected States

plt.figure(figsize=(12,6))

filtered_data = covid_df[
    covid_df['State/UnionTerritory'].isin(
        ['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh']
    )
]

ax = sns.lineplot(
    data=filtered_data,
    x='Date',
    y='Active_Cases',
    hue='State/UnionTerritory'
)

ax.set_title("Top 5 Affected States in India", fontsize=16)
plt.tight_layout()
plt.show()
```
